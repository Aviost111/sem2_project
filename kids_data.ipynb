```python
def raise_hand(x,y):
    count=0
    raised_hands=[]
    for (colname,colval) in x.iteritems():
        if(colname=='Time'):
            for i in range(1,y):
                if(colval.values[i]>=colval.values[i-1]+0.1):
                    count=count+1
    return count
    
```

```python
def raise_hand_i(x,y):
    count=0
    raised_hands=[]
    for (colname,colval) in x.iteritems():
        if(colname=='Time'):
            for i in range(1,y):
                if(colval.values[i]>=colval.values[i-1]+0.1):
                    count=count+1
                    raised_hands.append(i-1)     
    return raised_hands
```

```python
def bounds(x0,xlist,ylist,m,b,i_of_hands):
    i=1
    if(m==math.inf):
        while(xlist[x0+i]<=xlist[x0]+50 and xlist[x0+i]>=xlist[x0]-50):
            if (x0+i==len(xlist)-1 or (x0+i) in i_of_hands ):
                return x0+i
            i+=1
        return x0+i
    while(ylist[x0+i]<=xlist[x0+i]*m+b+120 and ylist[x0+i]>=xlist[x0+i]*m+b-120):
        if (x0+i==len(xlist)-1 or (x0+i) in i_of_hands ):
            break
        i+=1
    return x0+i
```
